```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Class=LabelEncoder()
```

```python
df['Class_n']=le_Class.fit_transform(df['DEATH_EVENT'])
```

```python
independent = df.drop(['DEATH_EVENT','Class_n'],axis='columns')
dependent = df['Class_n']
dependent
```

```python
independent
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=independent
y=dependent
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn import tree
model =tree.DecisionTreeClassifier()
```

```python
model=model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train, y_train)
```

```python
tree.plot_tree(model)
```

```python
import numpy as np
importances = model.feature_importances_
```

```python
feature_names=x.columns
import pandas as pd
model_importances = pd.Series(importances, index=feature_names)
model_importances
```

```python
model_importances.plot(kind='bar')
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
#accuracy
import numpy as np
print("Mean sum of squares (MSE): %.2f" % np.mean(((y_pred - y_test)** 2)**0.5))
```
